```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# freq = 1, mean = 0, std = 0.3031
# freq = 5, mean = 0.2, std = 0.22009
# freq = 10, mean = 0.45, std = 0.14827
# freq = 20, mean = 0.87, std = 0.0985
# freq = 30, mean = 1.28, std = 0.079

# sample 2000 points from each of the distributions and plot only the kde of them (instead of hist)
# freq is the name, mean is the mean, std is the standard deviation

stds = [0.02378, 0.05701, 0.07991, 0.12446, 0.15130]
means = [0.02, 0.038, 0.04, 0.09, 0.087]
freqs = [1, 5, 10, 20, 30]

data = {}
for i in range(5):
    data[freqs[i]] = np.random.normal(means[i], stds[i], 1000)
    
# plot the kde of the data
for key in data.keys():
    sns.kdeplot(data[key], label=f"Freq = {key}, std = {stds[freqs.index(key)]}")
plt.title("Relative P&L under Black-Scholes")
plt.xlabel("Relative P&L")
plt.ylabel("Density")
plt.legend()
plt.savefig("kde_BS_notransact.png")
plt.show()



```
